```python
top_10_countries= athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#Age Distribution
plt.figure(figsize=(12,6))
plt.title('Age Distribution Table')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athlete_df.Age, bins = np.arange(10,80,2),color='orange',edgecolor='white');
```

```python
female_participant=athlete_df[(athlete_df.Sex=="F") & (athlete_df.Season=='Summer')][['Sex','Year']]
female_participant=female_participant.groupby('Year').count().reset_index()
female_participant.tail()
```

```python
womanOlympics=athlete_df[(athlete_df.Sex=="F") & (athlete_df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womanOlympics,palette='Spectral')
plt.title('Woman Participation')
```

```python
part=womanOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Participation over time')
```
